```python
pairs = list(df['Pair'])
new_pairs = []


for pair in pairs: 
    new_pair = pair.split('/')[0] + pair.split('/')[1]
    new_pairs.append(new_pair)
    
btc_klines = client.get_historical_klines( symbol='BTCUSDT', interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
eth_klines = client.get_historical_klines( symbol='ETHUSDT', interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
bnb_klines = client.get_historical_klines( symbol='BNBUSDT', interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
xrp_klines = client.get_historical_klines( symbol='XRPUSDT', interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
```

```python
kline_array = []



for index, pair in enumerate(new_pairs):
    print(pairs[index])
    if pairs[index].split('/')[1] == 'BTC':
        kline_dict = {}
        klines = client.get_historical_klines( symbol=pair, interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
        total_order_volume = 0
        for kline_index, kline in enumerate(klines):
            total_order_volume += float(kline[7])* float(btc_klines[kline_index][4])
        kline_dict['average_volume'] = total_order_volume / len(kline)
        kline_dict['pair'] = pairs[index]
        kline_array.append(kline_dict)
    elif pairs[index].split('/')[1] == 'ETH':
        kline_dict = {}
        klines = client.get_historical_klines( symbol=pair, interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
        total_order_volume = 0
        for kline_index, kline in enumerate(klines):
            total_order_volume += float(kline[7])* float(eth_klines[kline_index][4])
        kline_dict['average_volume'] = total_order_volume / len(kline)
        kline_dict['pair'] = pairs[index]
        kline_array.append(kline_dict)
    elif pairs[index].split('/')[1] == 'XRP':
        kline_dict = {}
        klines = client.get_historical_klines( symbol=pair, interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
        total_order_volume = 0
        for kline_index, kline in enumerate(klines):
            total_order_volume += float(kline[7])* float(xrp_klines[kline_index][4])
        kline_dict['average_volume'] = total_order_volume / len(kline)
        kline_dict['pair'] = pairs[index]
        kline_array.append(kline_dict)
        
    elif pairs[index].split('/')[1] == 'BNB':
        kline_dict = {}
        klines = client.get_historical_klines( symbol=pair, interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
        total_order_volume = 0
        for kline_index, kline in enumerate(klines):
            total_order_volume += float(kline[7])* float(bnb_klines[kline_index][4])
        kline_dict['average_volume'] = total_order_volume / len(kline)
        kline_dict['pair'] = pairs[index]
        kline_array.append(kline_dict)

    else:
        kline_dict = {}
        klines = client.get_historical_klines( symbol=pair, interval=Client.KLINE_INTERVAL_1DAY, start_str="10 day ago UTC")
        total_order_volume = 0
        for kline in klines:
            total_order_volume += float(kline[7])
        kline_dict['average_volume'] = total_order_volume / len(kline)
        kline_dict['pair'] = pairs[index]
        kline_array.append(kline_dict)
```
